### The below problem "A histogram showing the count of each sentiment for the “Delta” airline"

#how I approached the problem was by creating two data frames. One grabbing the rows that had "Delta" in them and the other grabbing the values of airline_sentiment for the delta rows. I used pandas loc with a condition looking for "Delta" in the airline column. Then I made that list to a df. Then I grabbed the rows with pandas loc that looked for the column airline that had the value of "Delta". This table gave me the opportunity to grab all the values in airline_sentiment that were directed towards "Delta". Then I reset the index values of both tables and used pandas to make them into DataFrames. I had two DataFrames, one containing "Delta" as a value from the inital larger df and another with all the sentiment towards 'Delta'. I used the pandas concat method to combine the values, with axis=1 representing the columns. I then dropped index values.

```python
delta = df.loc[df.airline == "Delta", 'airline'].tolist()
```

```python
df.info()
```

```python
#grabbing all the rows with 'Delta' in the 'airline' column
airline_df = df.loc[df['airline'] == 'Delta']
```

```python
airline_df.head()
```

```python
# This needs to be made into a data frame with the sentiment regarding Delta
a_sentiment = airline_df.airline_sentiment
```

```python
# Now this is a data frame with the sentiment regarding Delta
a_sentiment_df = pd.DataFrame(a_sentiment)
```

```python
a_sentiment_df.head()
```

```python
a_sentiment_df.info()
```

```python
#dataframe that soley has delta as a value that lines up with the airline df 
df_delta = pd.DataFrame(delta)
```

```python
df_delta.head()
```

```python
df_delta = df_delta.reset_index()
```

```python
df_delta.head()
```

```python
a_sentiment_df = a_sentiment_df.reset_index()
```

```python
a_sentiment_df.head()
```

```python
#this produced the wrong data, this was just the sentiment aligned incorrectly with the df_delta 
#df_pop = df_delta[0] + ' ' + df['airline_sentiment']

#This should be the correct data frame
df_pop = pd.concat([df_delta, a_sentiment_df], axis=1)
```

```python
df_pop.head()
```

```python
df_pop.info()
```

```python
# I dropped the index and 0 column that I itially got from the concat to get a hisogram that solely showed the sentiment about Delta
#drop 0 and 'index'
df_pop = df_pop.drop('index', axis=1)
```

```python
df_pop.info()
```

```python
fig0 = px.histogram(df_pop)
```

```python
fig0.show()
```

```python
fig1 = px.histogram(df_pop, x="airline_sentiment", title='Sentiment towards Delta')
```

```python
fig1.show()
```
